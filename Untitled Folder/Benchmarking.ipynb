# Benchmarking
This notebook benchmarks multiple comparison-based optimization (CBO) algorithms on three test functions.


```python
import random
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('..')
from ExampleCode.benchmarkfunctions import SparseQuadratic, MaxK, NonSparseQuadratic
from Algorithms.gld_optimizer import GLDOptimizer
from Algorithms.stp_optimizer import STPOptimizer
from ExampleCode.oracle import Oracle
from Algorithms.SignOPT2 import SignOPT

import copy as copy
```

## Choose Function
Choose the benchmarking function and set it's parameters

```python
n = 5000
noiseamp = 0.001
num_trials = 3
num_queries = int(1e4)
obj_func = NonSparseQuadratic(n, noiseamp)
x_0_all = np.random.randn(n)
function_budget = 1000
direction_vector_type = 2


# initialize lists
stp_func_list = []
gld_func_list = []
signOPT_func_list = []

# Step_sizes
## It may make sense to have different step-sizes for different algs
step_size_STP = 1.0
step_size_signOPT = 0.1
```

```python
for trials in range(num_trials):
    # Choose new random initialization for each run
    x_0 = np.random.randn(n)
    
    # Do STP
    stp = STPOptimizer(direction_vector_type, x_0, n, step_size_STP, obj_func, function_budget)
    termination=False
    while termination is False:
        # optimization step
        solution, curr_func_value, termination = stp.step()
        print("current function value is {:+.4f}".format(curr_func_value))
    stp_func_list.append(stp.f_vals)
    
    # Now try SignOPT
    oracle = Oracle(obj_func)
    m = 10
    r = 0.001
    termination=False
    signOPT = SignOPT(oracle, function_budget, x_0, m, step_size_signOPT, r, debug=True, function=obj_func)
    while termination is False:
        # optimization
        solution, curr_func_value, termination = signOPT.step()
        print("current function value is {:+.4f}".format(curr_func_value))
        
    signOPT_func_list.append(signOPT.f_vals)
    
    # Now try GLD
    x_0 = np.random.randn(n)
    R_ = 10
    r_ = .01
    gld = GLDOptimizer(obj_func, x_0, R_, r_, function_budget)
    termination = False
    while termination is False:
        solution, curr_func_value, termination = gld.step()
        print("current function value is {:+.4f}".format(curr_func_value))
    
    gld_func_list.append(gld.f_vals)
```

```python
mean_STP = np.mean(stp_func_list, axis=0)
std_dev_STP = np.std(stp_func_list, axis=0)

mean_signOPT = np.mean(signOPT_func_list, axis=0)
std_dev_signOPT = np.std(signOPT_func_list, axis=0)

mean_GLD = np.mean(gld_func_list, axis=0)
std_dev_GLD = np.std(gld_func_list, axis=0)
```

```python
plt.plot(mean_STP, label="STP method")
plt.plot(mean_signOPT, label="Sign OPT")
plt.plot(mean_GLD, label="GLD")
plt.legend()
plt.show()
```
